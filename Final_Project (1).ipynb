```python
# Inventory Management System
inventory = []

# Function to add item
def add_item():
    name = input("Enter product name: ")
    quantity = int(input("Enter quantity: "))
    price = float(input("Enter price (Rs.): "))
    category = input("Enter category: ")

    item = {
        "name": name,
        "quantity": quantity,
        "price": price,
        "category": category
    }

    inventory.append(item)
    print(f"âœ… '{name}' added successfully!\n")

# Function to show all items
def show_inventory():
    if not inventory:
        print("ğŸ“­ Inventory is empty.\n")
        return
    print("ğŸ“¦ Inventory List:")
    for i, item in enumerate(inventory, 1):
        print(f"{i}. Name: {item['name']}, Qty: {item['quantity']}, Price: Rs.{item['price']}, Category: {item['category']}")
    print()

# Function to search item
def search_item():
    keyword = input("ğŸ” Enter name or category to search: ").lower()
    found = False
    for item in inventory:
        if keyword in item['name'].lower() or keyword in item['category'].lower():
            print(f"â¡ï¸  Name: {item['name']}, Qty: {item['quantity']}, Price: Rs.{item['price']}, Category: {item['category']}")
            found = True
    if not found:
        print("âŒ No item found with that keyword.\n")
    print()

# Function to update item
def update_item():
    name = input("âœï¸ Enter name of item to update: ").lower()
    for item in inventory:
        if item['name'].lower() == name:
            print("What do you want to update?")
            print("1. Quantity")
            print("2. Price")
            print("3. Category")
            choice = input("Enter choice (1/2/3): ")

            if choice == "1":
                item['quantity'] = int(input("Enter new quantity: "))
            elif choice == "2":
                item['price'] = float(input("Enter new price: "))
            elif choice == "3":
                item['category'] = input("Enter new category: ")
            else:
                print("âŒ Invalid choice.\n")
                return

            print("âœ… Item updated successfully!\n")
            return
    print("âŒ Item not found.\n")

# Function to delete item
def delete_item():
    name = input("ğŸ—‘ï¸ Enter item name to delete: ").lower()
    for item in inventory:
        if item['name'].lower() == name:
            inventory.remove(item)
            print(f"âœ… '{name}' removed from inventory.\n")
            return
    print("âŒ Item not found.\n")

# Function to calculate total inventory value
def total_value():
    total = 0
    for item in inventory:
        total += item['quantity'] * item['price']
    print(f"ğŸ’° Total Inventory Value = Rs.{total}\n")

# Menu function
def menu():
    while True:
        print("ğŸ“‹ MENU")
        print("1. Add Item")
        print("2. View Inventory")
        print("3. Search Item")
        print("4. Update Item")
        print("5. Delete Item")
        print("6. Inventory Value")
        print("0. Exit")
        choice = input("ğŸ‘‰ Choose an option: ")

        if choice == "1":
            add_item()
        elif choice == "2":
            show_inventory()
        elif choice == "3":
            search_item()
        elif choice == "4":
            update_item()
        elif choice == "5":
            delete_item()
        elif choice == "6":
            total_value()
        elif choice == "0":
            print("ğŸ‘‹ Exiting program. Goodbye!")
            break
        else:
            print("âŒ Invalid choice, try again.\n")

# Start the program
menu()
```
